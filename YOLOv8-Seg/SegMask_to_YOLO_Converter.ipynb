```python
import os
import cv2
import numpy as np

# Directories
input_dir = "SegData/masks"
output_dir = "SegData/processed_masks"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Process each mask
for mask_name in os.listdir(input_dir):
    if mask_name.endswith(".png"):
        mask_path = os.path.join(input_dir, mask_name)
        
        # Read the mask
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        
        # Convert 255 to 1
        mask = np.where(mask == 255, 1, mask)
        
        # Save the processed mask
        output_path = os.path.join(output_dir, mask_name)
        cv2.imwrite(output_path, mask)

print("Masks have been preprocessed and saved to:", output_dir)
```
